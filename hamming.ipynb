```python
#Hamming
def parity_bits(length):
    j=0
    pow2=[1,2,4,8,16,32]
    par=[]
    for i in range(0,length):        #for finding index of parity bits.
        if i+1 in pow2:
            par.append(i+1) 
    return par
def sender(data,parity,r):
    m=len(data)    
    elements=[]
    j=len(data)-1
    pow2=[1,2,4,8,16,32]
    par=parity_bits(m+r)
    for i in range(0,m+r): 
        if i+1 in pow2: 
            elements.append(i+1)
        else:
            elements.append(int(data[j]))
            j-=1
    j=1
    for i in par:
        string=""
        string+=str(i)
        for k in range(0,len(elements)+1):
            binary=""
            if k>i:
                binary=bin(k)
                if binary[-j]=='1':
                    string+=str(elements[k-1])
        j+=1
        count=0    #For finding parity Bits
        for m in range (1, len(string)):
            if string[m]=='1':
                count+=1
        if parity==1:
            if count%2==0:
                chng_par=list(string)
                chng_par[0]='0'
                string=''.join(chng_par)
            else:
                chng_par=list(string)
                chng_par[0]='1'
                string=''.join(chng_par)
        else:
            if count%2==0:
                chng_par=list(string)
                chng_par[0]='1'
                string=''.join(chng_par)
            else:
                chng_par=list(string)
                chng_par[0]='0'
                string=''.join(chng_par)#         print("P"+str(i)+"="+string[0])
        elements[i-1]=int(string[0])
    return elements

def receiver(hamming,parity,r):
    print("\n\t\tReceiver Side")
    par=parity_bits(len(hamming))
    elements=list(hamming)
    j=1
    if elements[3]=='0':#for inducing Error
        elements[3]='1'
    else:
        elements[3]='0'
    print("Received Hamming Code="+''.join(elements)[::-1])
    cal_detect=""
    for i in par:
        string=""
        for k in range(0,len(elements)+1):
            binary=""
            binary=bin(k).replace("0b","0"*10)
            if binary[-j]=='1':
                string+=str(elements[k-1])
        j+=1
        count=0    #For finding parity Bits
        for m in range (0, len(string)):
            if string[m]=='1':
                count+=1
        if parity==1:
            if count%2==0:
                cal_detect+='0'
            else:
                cal_detect+='1'
        else:
            if count%2==0:
                cal_detect+='1'
            else:
                cal_detect+='0'
    cal_detect=cal_detect[::-1]
    check=int(cal_detect,2)
    data=""
    if(check==0):
        print("No Errors detected\nData Accepted")
        for i in range(0,len(elements)):
            if i+1 not in par:
                data+=elements[i]
    else:
        print("Error Detected at position (from right)=",str(check))
        if elements[check-1]=='0':
            elements[check-1]='1'
        else:
            elements[check-1]='0'
        for i in range(0,len(elements)):
            if i+1 not in par:
                data+=elements[i]
        print("Corrected Hamming Code=",''.join(elements)[::-1])
    print("Data Received="+data[::-1])     
print("\n\t\tSender Side")
data=input("Enter the data=>")
p=int(input("Enter the parity in which data to be encoded [1:Even\t0:Odd] =>"))
m=len(data)
r=0
while(m+1>2**r-r):
    r+=1
print('Number of Parity Bits=',r)
codeword=sender(data,p,r)
for x in range(len(codeword)):
    codeword[x]=str(codeword[x])
codeword=''.join(codeword)
print("Sent Hamming Code=",codeword[::-1])
receiver(codeword,p,r)
```
